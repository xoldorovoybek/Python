Task: JSON Modification
Write a program that allows users to add new books, update existing book information, and delete books from the books.json JSON file.

```python
import json
import os

class BookManager:
    def __init__(self, filename='books.json'):
        self.filename = filename
        self.books = self.load_books()

    def load_books(self):
        """Load books from the JSON file, or return empty list."""
        if os.path.exists(self.filename):
            with open(self.filename, 'r') as f:
                return json.load(f)
        else:
            return []

    def save_books(self):
        """Save current books to JSON file."""
        with open(self.filename, 'w') as f:
            json.dump(self.books, f, indent=4)

    def add_book(self):
        """Add a new book."""
        title = input("Enter book title: ")
        author = input("Enter author name: ")
        year = input("Enter publication year: ")

        book = {
            'title': title,
            'author': author,
            'year': year
        }

        self.books.append(book)
        self.save_books()
        print(f"Book '{title}' added successfully.")

    def update_book(self):
        """Update existing book info."""
        title = input("Enter the title of the book to update: ")

        for book in self.books:
            if book['title'].lower() == title.lower():
                print(f"Current info: {book}")
                new_title = input("New title (Enter to skip): ")
                new_author = input("New author (Enter to skip): ")
                new_year = input("New year (Enter to skip): ")

                if new_title:
                    book['title'] = new_title
                if new_author:
                    book['author'] = new_author
                if new_year:
                    book['year'] = new_year

                self.save_books()
                print(f"Book '{title}' updated successfully.")
                return

        print(f"Book '{title}' not found.")

    def delete_book(self):
        """Delete a book by title."""
        title = input("Enter the title of the book to delete: ")

        new_books = [book for book in self.books if book['title'].lower() != title.lower()]

        if len(new_books) == len(self.books):
            print(f"Book '{title}' not found.")
        else:
            self.books = new_books
            self.save_books()
            print(f"Book '{title}' deleted successfully.")

    def view_books(self):
        """Display all books."""
        if not self.books:
            print("No books found.")
        else:
            print("\n=== Book List ===")
            for idx, book in enumerate(self.books, start=1):
                print(f"{idx}. {book['title']} by {book['author']} ({book['year']})")

    def run(self):
        """Run the main menu loop."""
        while True:
            print("\n=== Book Manager ===")
            print("1. Add new book")
            print("2. Update existing book")
            print("3. Delete a book")
            print("4. View all books")
            print("5. Exit")

            choice = input("Select an option (1-5): ")

            if choice == '1':
                self.add_book()
            elif choice == '2':
                self.update_book()
            elif choice == '3':
                self.delete_book()
            elif choice == '4':
                self.view_books()
            elif choice == '5':
                print("Goodbye!")
                break
            else:
                print("Invalid option. Please select 1-5.")

if __name__ == '__main__':
    manager = BookManager()
    manager.run()
```

Create a program that asks users for a movie genre and recommends a random movie from that genre.

```python
import requests
import random

API_KEY = "YOUR_OMDB_API_KEY"  # BU YERGA O'Z KALITINGIZNI QO'YING

# Janrlar bo'yicha ba'zi mashhur filmlar
movie_pool = {
    "action": ["Mad Max: Fury Road", "Die Hard", "The Dark Knight", "John Wick"],
    "comedy": ["Superbad", "Step Brothers", "The Hangover", "Anchorman"],
    "drama": ["Forrest Gump", "Fight Club", "The Shawshank Redemption", "The Godfather"],
    "horror": ["The Conjuring", "It", "A Nightmare on Elm Street", "The Ring"],
    "sci-fi": ["Inception", "Interstellar", "The Matrix", "Blade Runner 2049"],
    "romance": ["Titanic", "The Notebook", "La La Land", "Pride and Prejudice"]
}

def get_movie(title):
    """OMDb API dan film ma'lumotini olib keladi"""
    url = "http://www.omdbapi.com/"
    params = {
        "apikey": API_KEY,
        "t": title
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data["Response"] == "True":
            return data
        else:
            print(f"Movie '{title}' not found.")
            return None
    else:
        print("Error fetching data.")
        return None

def recommend_movie():
    """Foydalanuvchidan janr so'raydi va random film tavsiya qiladi"""
    genre = input("Enter a movie genre (action, comedy, drama, horror, sci-fi, romance): ").lower()
    if genre not in movie_pool:
        print("Sorry, that genre is not available.")
        return

    movie_title = random.choice(movie_pool[genre])
    movie = get_movie(movie_title)

    if movie:
        print("\n🎬 Recommended Movie 🎬")
        print(f"Title: {movie['Title']}")
        print(f"Year: {movie['Year']}")
        print(f"Genre: {movie['Genre']}")
        print(f"Director: {movie['Director']}")
        print(f"Plot: {movie['Plot']}")
        print(f"IMDB Rating: {movie['imdbRating']}")
    else:
        print("No movie data available.")

if __name__ == "__main__":
    recommend_movie()
```
